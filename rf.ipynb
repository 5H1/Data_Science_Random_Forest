```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
# Identify numerical columns
df_num = ds[['RI', 'Na', 'Mg', 'Al', 'Si','K','Ca','Ba','Fe']]
```

```python
df_num['RI'] = SS.fit_transform(df_num[['RI']])
df_num['Na'] = SS.fit_transform(df_num[['Na']])
df_num['Mg'] = SS.fit_transform(df_num[['Mg']])
df_num['Al'] = SS.fit_transform(df_num[['Al']])
df_num['Si'] = SS.fit_transform(df_num[['Si']])
df_num['K'] = SS.fit_transform(df_num[['K']])
df_num['Ca'] = SS.fit_transform(df_num[['Ca']])
df_num['Ba'] = SS.fit_transform(df_num[['Ba']])
df_num['Fe'] = SS.fit_transform(df_num[['Fe']])

df_num.head()
```

```python

# Data partition
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
#2. Implement a Random Forest classifier using Python and a machine learning library like scikit-learn.
```

```python

#step6: model fitting
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()

model.fit(X_train,Y_train)
Y_pred_train = model.predict(X_train)
Y_pred_test  = model.predict(X_test)

from sklearn.metrics import accuracy_score
train_acc = accuracy_score(Y_train,Y_pred_train)
print("Training Accuracy:",train_acc)
test_acc= accuracy_score(Y_test,Y_pred_test)
print("Test Accuracy :",test_acc)
```

```python

#=========================================================
# cross validation with randomforest Classifier
#=========================================================
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

training_acc = []
test_acc = []

for i in range(100,201):
    X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.25,random_state=i)
    model = RandomForestClassifier(max_samples=0.8,max_features=0.7)
    model.fit(X_train,Y_train)
    Y_pred_train = model.predict(X_train)
    Y_pred_test  = model.predict(X_test)
    training_acc.append(accuracy_score(Y_train,Y_pred_train))
    test_acc.append(accuracy_score(Y_test,Y_pred_test))

import numpy as np
print("Cross validation - Training Accuarcy:" ,np.mean(training_acc).round(2))
print("Cross validation - Test Accuracy:" ,np.mean(test_acc).round(2))
print("RandomForestClassifier -variance :" ,(np.mean(training_acc)-np.mean(test_acc)).round(2))
```

```python

#=========================================================
# cross validation with randomforest Classifier
#=========================================================
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

training_acc = []
test_acc = []

for i in range(100,201):
    X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.25,random_state=i)
    model = RandomForestClassifier(max_depth=6, max_features='sqrt', max_leaf_nodes=6,max_samples=0.7, n_estimators=50)
    model.fit(X_train,Y_train)
    Y_pred_train = model.predict(X_train)
    Y_pred_test  = model.predict(X_test)
    training_acc.append(accuracy_score(Y_train,Y_pred_train))
    test_acc.append(accuracy_score(Y_test,Y_pred_test))

import numpy as np
print("Cross validation - Training Accuarcy:" ,np.mean(training_acc).round(2))
print("Cross validation - Test Accuracy:" ,np.mean(test_acc).round(2))
print("RandomForestClassifier -variance :" ,(np.mean(training_acc)-np.mean(test_acc)).round(2))
```
